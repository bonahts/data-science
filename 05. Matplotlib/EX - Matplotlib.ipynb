```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0, 100)
y = x*2
z = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('XLabel')
ax.set_ylabel('YLabel')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,.5,.2,.2])

ax1.plot(x,y)
ax2.plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
```

# Use x,y, and z arrays to create the plot below. Notice the xlimits and y limits on the inserted plot:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

#Large plot
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

#Insert 
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

# Now plot(x,y) and (x,z) on the axes.Play around with the linewidth and style

```python
# fig,axes = plt.subplots(nrows=1,ncols=2)
fig,axes = plt.subplots(1,2)
# lw = linewidth, ls = linestyle
axes[0].plot(x,y,color='blue',lw='3',ls='--')
axes[1].plot(x,z,color='red',lw='3')
```

```python
# fig,axes = plt.subplots(nrows=1,ncols=2)
fig,axes = plt.subplots(1,2,figsize=(12,2))
# lw = linewidth, ls = linestyle
axes[0].plot(x,y,color='blue',lw='3',ls='--')
axes[1].plot(x,z,color='red',lw='3', ls='--')
```
