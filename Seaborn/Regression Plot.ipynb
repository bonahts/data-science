```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'],
          scatter_kws=
          {'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,col='sex',row)
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,col='sex',row='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,col='time',row='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,col='day',hue='sex', 
           aspect=0.6, size=8)
```
